```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
nbastats=pd.read_csv('NBA Season Data.csv')
nbastats.head()
```

```python
nbastats=nbastats.rename(columns={"Year":"Season_End"})
```

```python
#nbastats=nbastats[nbastats["Yrs Experience"]>=2]
```

```python
nbastats["Season_End"]=nbastats.Season_End-1
```

```python
nbastats=nbastats[(nbastats.Season_End>=2000)]
```

```python
nbastats=nbastats.rename(columns={'Tm':"Team"})
```

```python
nbastats=nbastats[['Team','Player ID','Player','Season_End','Age','Age on Feb 1','G','MP','MPG','PER','ORB%','DRB%','AST%','STL%','BLK%','USG%','WS','WS/48','BPM','WORP','Adjusted WORP','Yrs Experience','Height','Weight','Rounded Position','Estimated Position','TrueSalary','Adjusted Production']]
```

```python
nbastats[nbastats.isnull().any(axis=1)]
```

```python
len(nbastats)-nbastats.count()
```

```python
nbastats.dropna(inplace=True)
```

```python
nbastats.columns
```

```python
nbastats['TrueSalary']=nbastats.TrueSalary.str.replace("$","").str.replace(",","").astype(float)
```

```python
from sklearn.cross_validation import cross_val_score
```

```python
feature_cols=['Age on Feb 1','BPM','PER','WS','USG%','MPG','STL%','DRB%','ORB%','Yrs Experience','AST%','BLK%']
X=nbastats[feature_cols]
y=nbastats.TrueSalary
```

```python
from sklearn import linear_model
lm=linear_model.LinearRegression()
scores=cross_val_score(lm,X,y,cv=10,scoring="r2")
print (scores.mean())
```
